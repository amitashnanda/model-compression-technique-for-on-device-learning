```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
rem_ratio = [0.7001855287569573, 0.6701298701298701, 0.6371057513914656, 0.6007421150278293, 0.561038961038961, 0.5168831168831168, 0.4686456400742115, 0.4155844155844156, 0.35695732838589983, 0.2927643784786642]
rem_parameters = [i*2736 for i in rem_ratio]
val_loss = [0.4029, 0.3570, 0.3590, 0.3790, 0.3737, 0.3831, 0.3690, 0.3768, 0.3620, 0.3792]
val_accuracy = [0.8203, 0.8437, 0.8491, 0.8540, 0.8573, 0.8582, 0.8605, 0.8601, 0.8649, 0.8554]
```

```python
rem_parameters = rem_parameters[::-1]
val_loss = val_loss[::-1]
val_accuracy = val_accuracy[::-1]
```

```python
data = {
    'rem_parameters': rem_parameters,
    'val_loss': val_loss,
    'val_accuracy': val_accuracy
}

df = pd.DataFrame(data)
```

### Total parameters are 2736

```python
sns.set(style="whitegrid")

# Creating a scatter plot for val_loss
plt.figure(figsize=(6, 4))
sns.lineplot(x='rem_parameters', y='val_loss', data=df, label='Validation Loss')

# Adding labels and title
plt.xlabel('Remaining Parameters')
plt.ylabel('Validation Loss')
plt.title('Remaining Parameters vs Validation Loss')
plt.gca().invert_xaxis()

# Adding legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Plotting Validation Accuracy with seaborn
plt.figure(figsize=(6, 4))
sns.lineplot(x='rem_parameters', y='val_accuracy', data=df, color='orange', label='Validation Accuracy')

# Adding labels and title
plt.xlabel('Remaining Parameters (Descending)')
plt.ylabel('Validation Accuracy')
plt.title('Remaining Parameters vs Validation Accuracy')
plt.gca().invert_xaxis()

# Adding legend
plt.legend()

# Show the plot
plt.show()
```
